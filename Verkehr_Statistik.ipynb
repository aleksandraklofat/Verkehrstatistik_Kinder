# Kinder in Strassenunfällen in Deutschland 

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# downloading data
df = pd.read_csv('verkehr_stat.csv',sep=';', encoding='latin-1')
```

```python
df.head(10)
```

## Getötete Kinder unter 15 auf deutschen Strassen: Gesamtstatistik

```python
gesamt_stat = df[df['Art'] == 'Insgesamt'] # Ausfiltern von Fußgänger
gesamt_get = gesamt_stat[gesamt_stat['Statistik'] == 'Getötete'] #Ausfiltern von gesamten Unfällen
```

```python
gesamt_get
```

### Visualisierung in Matplotlib

```python
plt.plot('Jahr', 'Zahl', data=gesamt_get)
plt.xlabel('Jahr')
plt.ylabel('Kinder unter 15')
plt.title('Getötete Kinder unter 15 im Straßenverkehr')
```

## Getötete Kinder unter 15 auf deutschen Strassen: Fußgänger

```python
# Wir filtern den Datensatz nach Fußgänger
```

```python
gesamt_fusg = df[df['Art'] == 'Fußgänger'] # Ausfiltern von Gesamtangaben
df_fusg = gesamt_fusg[gesamt_fusg['Statistik'] == 'Getötete'] #Ausfiltern von gesamten Unfällen
```

```python
df_fusg
```

### Visualisierung in Matplotlib

```python
plt.plot('Jahr', 'Zahl', data=df_fusg)
plt.xlabel('Jahr')
plt.ylabel('Füßgänger Kinder unter 15')
plt.title('Getötete Kinder im Straßenverkehr (Fußgänger)')
```

## Kinder in Deutschland

```python
plt.plot('Jahr', 'Kinder_ges', data=df)
plt.xlabel('Jahr')
plt.ylabel('Zahl der Kinder')
plt.title('Kinder unter 15 in Deutschland')
```

## Tödliche Unfälle im Verheltnis zur Gesamtzahl der Kinder in DE

```python
gesamt_get #getötete Kinder unter 15 im Strassenverkehr
```

Wir setzen die Zahlen in Relation:

```python
get_rel = gesamt_get['Zahl']/gesamt_get['Kinder_ges'] #Zahl der getöteten geteilt durch die Gesamtzahl der Kinder
```

```python
get_rel
```

```python
df_rel_t = pd.concat([gesamt_get,get_rel], axis=1) #Aufnehmen der Variable in den Datensatz
```

```python
df_rel_t
```

```python
df_rel_t = df_rel_t.rename(columns = {0: "Relation"}) 
```

```python
df_rel_t
```

```python
plt.plot('Jahr', 'Relation', data=df_rel_t)
plt.xlabel('Jahr')
plt.ylabel('Relative Zahlen: Kinder unter 15')
plt.title('Getötete Kinder im Straßenverkehr (relativ zur Gesamtpopulation)')
```

## Verhältnis zu Kinder als Fussgänger

```python
df_fusg.head(20)
```

```python
relation_variable = df_fusg['Zahl']/df_fusg['Kinder_ges']
```

```python
df_fusg_new = pd.concat([df_fusg,relation_variable], axis=1)
```

```python
df_fusg_new
```

```python
df_fusg_rel = df_fusg_new.rename(columns = {0: "Relation_Fusg"}) 
```

```python
df_fusg_rel
```

```python
plt.plot('Jahr', 'Relation_Fusg', data=df_fusg_rel)
plt.xlabel('Jahr')
plt.ylabel('Relative Zahlen: Kinder unter 15')
plt.title('Getötete Kinder (Fußgänger) (relativ zur Gesamtpopulation)')
```
